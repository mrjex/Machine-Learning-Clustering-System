```python
# Returns a list of the contained categorical features
def get_categorical_features():
  object_types = (data.dtypes == 'object')
  return list(object_types[object_types].index)

# Perform Label Encoding to transform the features into numbers
def numerical_label_encoding(categorical_features):
  LE = LabelEncoder()
  for i in categorical_features:
    data[i] = LE.fit_transform(data[i])
  return data
```

```python
#######    BACKEND CLUSTERING     ##########

clustering_methods = {
    "K-Means": KMeans(n_clusters=num_clusters, random_state=42),
    "Agglomerative": AgglomerativeClustering(n_clusters=num_clusters),
    "Spectral": SpectralClustering(n_clusters=num_clusters, random_state=42, affinity='nearest_neighbors')
}

# Agglomerative Clustering
yhat_AC = clustering_methods["Agglomerative"].fit_predict(PCA_ds)
PCA_ds["Clusters_AC"] = yhat_AC

# K-Means Clustering
yhat_kmeans = clustering_methods["K-Means"].fit_predict(PCA_ds.iloc[:, :NUM_PCA_COMPONENTS])
PCA_ds["Clusters_KMeans"] = yhat_kmeans

# Spectral Clustering
yhat_spectral = clustering_methods["Spectral"].fit_predict(PCA_ds.iloc[:, :NUM_PCA_COMPONENTS])
PCA_ds["Clusters_Spectral"] = yhat_spectral
```

```python
#######    FRONTEND VISUALIZATION     ##########

# Create a color array for each clustering method
colors_ac = [pal[label] for label in PCA_ds["Clusters_AC"]]
colors_kmeans = [pal[label] for label in PCA_ds["Clusters_KMeans"]]
colors_spectral = [pal[label] for label in PCA_ds["Clusters_Spectral"]]

fig = plt.figure(figsize=(15, 5))

# Agglomerative Clustering
ax = fig.add_subplot(131, projection=f'{NUM_PCA_COMPONENTS}d')
ax.scatter(PCA_ds["col1"], PCA_ds["col2"], PCA_ds["col3"], c=colors_ac)
ax.set_title("Agglomerative Clustering")

# K-Means Clustering
ax = fig.add_subplot(132, projection=f'{NUM_PCA_COMPONENTS}d')
ax.scatter(PCA_ds["col1"], PCA_ds["col2"], PCA_ds["col3"], c=colors_kmeans)
ax.set_title("K-Means Clustering")

# Spectral Clustering
ax = fig.add_subplot(133, projection=f'{NUM_PCA_COMPONENTS}d')
ax.scatter(PCA_ds["col1"], PCA_ds["col2"], PCA_ds["col3"], c=colors_spectral)
ax.set_title("Spectral Clustering")

plt.show()
```

**Insights:** Each clustering method has different outcomes, yet they are quite similar in their partitioning, such that similar characteristics of the groups can be spotted. Thus, their cluster performance, or silhouette scores can be expected to be

**Conclusions:** Despite the differences of algorithmic approaches for each method, they produce relatively similar results for the particular dataset in this project. For Agglomerative, K-Means and Spectral Clustering, the parameter that is common for all is K. In essence, this parameter allows developers to adjust the number of clusters that the method produces. In this way, the depth or thoroughness that the method has to perform in its analysis of clusters is regulated. A higher value for K suggests that a more detailed search of feature discrepancies and relationships has to be conducted, as each data partition must be distinguished into separate clusters by their qualities. Conversely, lower values for K entail a more simple division of data points, where only the most prominent feature relationships that are easy to locate divides the few clusters. Thus, the very parameter K bridges the Agglomerative, K-Means and Spectral Clustering, which, as the visualization above illustrates, results in cases where their performances are similar if their configured K values are identical. Conclusively, K serves as the common denominator for these methods, making them applicable to problems of similar nature
